```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tau = 100
k = 1/200 #np.pi/2 * 1/tau # 
def func(f):
    return k**2+f**2-2*f*k*np.sin(f*tau)

def a_func(f):
    z = f*tau
    dz = z-np.pi/2
    return k**2+1/tau**2*np.pi**2/4+np.pi*dz/tau**2+dz**2/tau**2-2*k/tau*(np.pi/2+dz-np.pi/4*dz**2-dz**3/2)

def get_f(tau,k):
    return 1/tau*(np.pi/3-1/3*1/(tau*k)+np.sqrt((1/3*1/(k*tau)+np.pi/6)**2-1/3*np.pi/(k*tau)+2/3))
```

```python
c_k = 1/(np.pi/2 * 1/tau)
c_tau = 1/k*np.pi/2
print(c_k, c_tau)
```
