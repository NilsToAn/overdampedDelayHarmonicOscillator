```python
import numpy as np
import matplotlib.pyplot as plt
from setup import setup_matplotlib, one_col, two_col, std_height

setup_matplotlib()
```

```python
times = np.array([0, 1, 2, 3, 4, 5])
data = np.array([1, 2, 4, 3, 2, 3])

time_ticks = [r"$i-\tau_n$", r"$i-\tau_n+1$", "", "", r"$i$", r"$i+1$"]

i_idx = np.array([0, 1, len(times) - 2, len(times) - 1])

i_times = times[i_idx]
i_data = data[i_idx]
```

```python
def create_dim_line(ax, start, end, text, oft=np.array([0, -1]), rotation=0):
    ax.annotate(
        "",
        xy=start,
        xytext=end,
        arrowprops={
            "arrowstyle": "|-|",
            "shrinkA": 0,
            "shrinkB": 0,
            "color": "k",
            "mutation_scale": 2,
        },
    )
    ax.text(
        *((start + end) / 2 + oft),
        text,
        rotation=rotation,
        horizontalalignment="center",
        verticalalignment="center",
    )


fig, ax = plt.subplots()

ax.plot(times, data, "o-", alpha=0.5)
ax.plot(i_times, i_data, "o")

xlim = ax.set_xlim(*ax.set_xlim())
ylim = ax.set_ylim(-0.5, 5.5)

create_dim_line(
    ax,
    np.array([0, 0.5]),
    np.array([4, 0.5]),
    r"$\boldsymbol{s}^i$",
    oft=np.array([0, -0.35]),
)
create_dim_line(
    ax,
    np.array([1, 4.5]),
    np.array([5, 4.5]),
    r"$\boldsymbol{s}^{i+1}$",
    oft=np.array([0, 0.3]),
)

ax.set_xticks(times, time_ticks)
ax.set_yticks(data, data)


ax.set_xlabel("Temporal index")
ax.set_ylabel("Spatial index $l$")
plt.tight_layout()
plt.savefig("graphics/state_vis.pdf")
```
